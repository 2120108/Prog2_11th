# 問題1（初級）

以下の行列に逆行列がある場合その行列を求めましょう
$$
A=\left[
\begin{matrix} 1 & 5 \\ 4 & -22 \end{matrix}
\right]
$$


根岸　花帆

* 問題提案者の氏名を問題末尾に記入して下さい．複数人で作成した場合は連名で構いません．

```python
# ここに解答または解答例を記入する
import numpy as np

A = np.array([[1,5],[4,-2]])
print(A)

#行列式が0でない場合逆行列が存在する
#行列式を求める

a = np.linalg.det(A) 
print(a)

#逆行列を求める
A_inv = np.linalg.inv(A) 
print(A_inv)
```
